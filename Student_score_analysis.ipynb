```python
df = pd.read_csv('student_score.csv')
df = df.drop("Unnamed: 0", axis=1)
df = df.drop("TransportMeans", axis=1)
df
```

```python
df['WklyStudyHours'] = df['WklyStudyHours'].str.replace("05-Oct","5-10")
df
```

```python
df.isnull().sum()
```

# Analysis

```python
plt.figure(figsize=(5,5))
ax =sns.countplot(data = df,x= "Gender")
ax.bar_label(ax.containers[0])
plt.show()
```

```python
#female data is more them male data
#parent education impact on student score
```

```python
gn = df.groupby("ParentEduc").agg({"MathScore":'mean',"ReadingScore":"mean","WritingScore":"mean"})

print(gn)
```

```python
plt.figure(figsize = (5,5))
sns.heatmap(gn,annot = True)
plt.title("Relationship between Parents Education and Student's Score")
plt.show()
```

# Distribution of Ethnic Groups

```python
groupA = df.loc[(df['EthnicGroup']=="group A")].count()
groupB = df.loc[(df['EthnicGroup']=="group B")].count()
groupC = df.loc[(df['EthnicGroup']=="group C")].count()
groupD = df.loc[(df['EthnicGroup']=="group D")].count()
groupE = df.loc[(df['EthnicGroup']=="group E")].count()
l = ["group A","group B","group C","group D","group E"]
m = [groupA["EthnicGroup"],groupB["EthnicGroup"],groupC["EthnicGroup"],groupD["EthnicGroup"],groupE["EthnicGroup"]]
plt.pie(m,labels=l,autopct = "%1.2f%%")
plt.title("Distribution of Ethnic Group")
plt.show()
```
